```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# =====================================================
# 1️⃣ quotes 1~10페이지 크롤링
# =====================================================
base_url = "http://quotes.toscrape.com/page/{}/"

quotes_list = []

for page in range(1, 11):
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all("span", class_="text")

    for q in quotes:
        text = q.get_text()
        quotes_list.append(text)

print(f"총 quotes 수: {len(quotes_list)}")

# =====================================================
# 2️⃣ CountVectorizer (단어 벡터화)
# =====================================================
vectorizer = CountVectorizer(stop_words='english')  # 불용어 제거
X = vectorizer.fit_transform(quotes_list)

# 단어 리스트
words = vectorizer.get_feature_names_out()

# =====================================================
# 3️⃣ DataFrame 생성 (사진처럼)
# =====================================================
df = pd.DataFrame(X.toarray(), columns=words)

print("\n📊 CountVectorizer 결과 (일부):")
print(df.head())

# =====================================================
# 4️⃣ 가장 많이 나온 단어 TOP 5
# =====================================================
word_counts = df.sum().sort_values(ascending=False)

top5 = word_counts.head(5)

print("\n🔥 가장 많이 등장한 단어 TOP 5")
print(top5)
```
