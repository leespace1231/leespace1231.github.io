```python
try:
    from faker import Faker
except ImportError:
    !pip install faker
    from faker import Faker

import random

fake = Faker()

class User:
    def __init__(self, name=None):
        # кё°ліё м •ліҙ (кіјм ң 1)
        self.name = name if name else fake.name()
        self.email = fake.email()
        self.address = fake.address()
        self.job = fake.job()

        # нҢ”лЎңмҡ° м •ліҙ (кіјм ң 2)
        self.following = 0
        self.followers = 0

    # нҢ”лЎңмҡ° кё°лҠҘ (Tomмқҙ Jerry м«“м•„к°Җкё°)
    def following_mode(self, other_user):
        self.following += 1
        other_user.followers += 1
        print(f"{self.name} вҶ’ {other_user.name} нҢ”лЎңмҡ°!")

    def __repr__(self):
        return f"{self.name} ({self.email})"


# =========================
# вң… кіјм ң 1: 1000лӘ… мғқм„ұ
# =========================
users = [User() for _ in range(1000)]

print("мғқм„ұлҗң мң м Җ мҳҲмӢң:")
for i in range(5):
    print(users[i])


# =========================
# вң… кіјм ң 2: Tom вҶ’ Jerry 10лІҲ нҢ”лЎңмҡ°
# =========================
tom = User("Tom")
jerry = User("Jerry")

print("\n[Tomмқҙ JerryлҘј м«“м•„к°ҖлҠ” мӨ‘]")
for _ in range(10):
    tom.following_mode(jerry)

print("\nкІ°кіј")
print(f"Tom following: {tom.following}")
print(f"Jerry followers: {jerry.followers}")
```
